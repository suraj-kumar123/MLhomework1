```python
for df in temp_train:
    for i in range (0, len(df)):
        if np.isnan(df["Age"][i]) == True:
            df["Age"][i] = np.random.uniform(low=df['Age'].std(), high=df['Age'].median())
```

```python
temp_train[0].head()
```

```python
for i in range(len(train_data['Age'])):
    if np.isnan(train_data["Age"][i]) == True:
        train_data.at[i, 'Age']=np.random.uniform(train_data['Age'].mean(),train_data['Age'].std())
```

```python
train_data.head()
```

```python

train_data['FareBand'] = pd.qcut(train_data['Fare'], 4)
train_data[['FareBand', 'Survived']].groupby(['FareBand']).mean()
val = train_data.FareBand.unique().get_values()
val.sort()
for df in temp_train:
    for i in range(len(val)):
        df.loc[(df['Fare'] > val[i].left) & (df['Fare'] <= val[i].right), 'Fare'] = i
    df['Fare'] = df['Fare'].astype(int)
```
